```python
df.drop('CODIGO_INTERNACAO', axis=1, inplace=True)
df.drop('DATA_INTERNACAO', axis=1, inplace=True)
df.drop('CODIGO_EVENTO', axis=1, inplace=True)
df.drop('EVENTO_OPERADORA', axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
df.rename(columns={'TITULARIDADE': 'IS_TITULAR',
                   'GENERO': 'IS_MASCULINO',
                   'FLAG_INTERNACAO': 'IS_INTERNACAO',
                   'REDE_REEMBOLSO': 'IS_REDE'}, inplace=True)

df['IS_TITULAR'] = df['IS_TITULAR'].map({'DEPENDENTE': False, 'TITULAR': True})
df['IS_MASCULINO'] = df['IS_MASCULINO'].map({'FEMININO': False, 'MASCULINO': True})
df['IS_REDE'] = df['IS_REDE'].map({'REEMBOLSO': False, 'REDE': True})
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df['EMPRESA'] = df['EMPRESA'].astype(str)
df['CODIGO_APOLICE'] = df['CODIGO_APOLICE'].astype(str)
df['CODIGO_SERVICO'] = df['CODIGO_SERVICO'].astype(str)
df['DATA_REFERENCIA'] = pd.to_datetime(df['DATA_REFERENCIA'], format='%Y-%m-%d')
df['DATA_ATENDIMENTO'] = pd.to_datetime(df['DATA_ATENDIMENTO'], format='%Y-%m-%d')
df['DATA_PAGAMENTO_EVENTO'] = pd.to_datetime(df['DATA_PAGAMENTO_EVENTO'], format='%Y-%m-%d')
```
